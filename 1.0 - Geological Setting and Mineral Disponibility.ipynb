### Introduction

The study area is located in the southwestern of Minas Gerais (MG) State, Brazil. It is one of the most important mining areas in Brazil, having operation of large mining companies such as Mosaic, Vale, Votorantim, Kinross, CBMM and Yara.

The area where the sediments samples where collected is inside a very rich regions in the mining point of view. Nearby cities like Araxá, Paracatu, Vazante and Catalão have several mines of different ores, such as Zn, Au, Ag and many others.

- According to Vinicius Xavier, the area has potential for Zn, Pb and Ag

- `Teles (2019)` says that Coromandel area has phosphate depositis in Retiro Fm, Vazante Group, with a P2O5 grade between 2 to 34%.

- `Dardenne et al. (1998)` confirm that the Zn and Pb deposits in the Paracatu-Vazante belt are associated with Vazante Group.

- `Oliveira (2011)` states that above Vazante Group rocks, there are metasediments belong to Canastra Group. This group is responsible for hosting gold deposits called Morro do Ouro.

- In the area might have kimberlites and lamproites. Even though geochemistry is not suited to regional sampling programmes, we'll try to use `Gregory and Tooms (1969, apud Muggeridge, 1995)` aproach, where they found anomalies approximately half kilometre of the body. **Nb, Ni, Cr and Mg were the most useful pathfinder elements**. Dispite this study, it's important to be aware due to the fact that might have a lot of ambiguity in results.

### Geological Setting

Coromandel, Patos de Minas, Patrocínio and Monte Carmelo are the nearest cities around the sampling area. 

This area is inside Brasilia Belt, and we can divide in 3 parts `(Fuck et al. (1993); Fuck et al. (1994))`:

- Brasilia Belt Internal Zone: Araxá Gr, Monte Carmelo and Abadia dos Dourados Complex

- Brasilia Belt External Zone: Canastra Gr and Ibiá Gr.

- Cratonic Zone: Vazante Gr, Paranoá Gr and Bambuí Gr.
